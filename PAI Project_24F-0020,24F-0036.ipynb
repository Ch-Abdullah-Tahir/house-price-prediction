```python
df=df.drop("sqft_above",axis=1)
display(df.head())
```

# Apply Log transformations


```python
df['log_price'] = np.log1p(df['price'])
df["log_sqft_living15"]=np.log1p(df['sqft_living15'])
df["log_sqft_living"]=np.log1p(df['sqft_living'])
df['log_sqft_basement']=np.log1p(df['sqft_basement'])
#for basement 
df['basement_exists'] = (df['sqft_basement'] > 0).astype(int)
df['log_sqft_basement'] = df['log_sqft_basement'].fillna(df['log_sqft_basement'].mean)
# Replaces 0 with mean so we don’t take log(0) (which is undefined).
# .apply(np.log)
# Takes the natural logarithm of all non-zero basement sizes.
```

# Apply one hot encoding

```python
df_encoded = df.copy()
#df_encoded = pd.get_dummies(df_encoded, columns=['bathrooms'], drop_first=True)
df_encoded = pd.get_dummies(df_encoded, columns=['view'], drop_first=True)
df_encoded = pd.get_dummies(df_encoded, columns=['floors'], drop_first=True)
#df_encoded = df_encoded.drop(columns=['view'], errors='ignore')
# df_encoded = pd.get_dummies(df_encoded, columns=["zipcode"], drop_first=True)
df_encoded.columns
```

```python
# Average log price per zipcode
zip_price_map = df.groupby("zipcode")["log_price"].mean()
#Replace the zip code with its corresponding mean log price:
df_encoded['zipcode_price'] = df_encoded['zipcode'].map(zip_price_map)
df_encoded.columns
```

```python
#Import and create the model
from sklearn.linear_model import LinearRegression
model=LinearRegression()
#this model will do the price prediction for us
model.fit(X_train,y_train)
#model will learn the features and target)
```

```python
y_pred=model.predict(X_test)
```

```python
#Evaluate the model
from sklearn.metrics import mean_squared_error,r2_score
MeanSquaredError=mean_squared_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
print("Mean Squared Error:",MeanSquaredError)
print("R-squared:", r2)
#mse=sse/n we divide it by n to make it independent of size of data set
#SSE IS TOTAL ERROR AND MSE IS AVERAGE ERROR
```

```python
# Mean Squared Error: 0.03454271650340142
# R-squared: 0.8788776544230653
```

```python
# For training data
r2_train = model.score(X_train, y_train)
print("Training R-squared:", r2_train)

# For testing data
r2_test = model.score(X_test, y_test)
print("Testing R-squared:", r2_test)
```
